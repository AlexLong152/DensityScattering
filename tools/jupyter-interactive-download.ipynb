```python
eps=0.001
srg = "lambdaSRGNN"
srgVal = 2.236
df=densdb.pddf
# full set of columns
# ['kind', 'c1', 'c3', 'c4', 'cd', 'ce', 'cE1', 'cE2', 'cE3', 'cE4', 'cE5', 'cE6', 'cE7', 'cE8', 'cE9', 'cE10', 'N', 'Z', 'Jtot', 'MTtot', 'j12max', 'lmax', 'lsummax', 'Nmax', 'OmegaHO', 'omega', 'theta', 'hashname', 'uniquefilename', 'localonly', 'localavail', 'addtime', 'moditime', 'E[MeV]', 'NP12A', 'NP12B', 'NP12', 'P12A', 'P12B', 'P12C', 'NP34A', 'NP34B', 'NP34', 'P34A', 'P34B', 'P34C', 'NP3A', 'NP3B', 'NP3', 'P3A', 'P3B', 'P3C', 'NQ4A', 'NQ4B', 'NQ4', 'Q4A', 'Q4B', 'Q4C', 'NQA', 'NQB', 'NQ', 'QA', 'QB', 'QC', 'NPTA', 'NPTB', 'NPT', 'PTA', 'PTB', 'PTC', 'NEA', 'NEB', 'NE', 'EA', 'EB', 'EC', 'NX', 'taumax', 'l3max', 'l4max', 'lammax', 'MODENN', 'orderNN', 'LambdaNN', 'potnr', 'empotnr', 'lam3NF', 'tnforder', 'j12max3nf', 'j3max3nf', 'relcalc', 'nnsrg', 'tnfsrg', 'lambdaSRGNN', 'lambdaSRG3N', 'potbareNN', 'potbareNNTNF', 'ostatSRGNN', 'ostatSRGTNF', 'cutnumSRGNN', 'cutnumSRGTNF', 'j12maxdens', 'taumaxdens', 'lmaxdens', 'nphitilde', 'nxtilde', 'wfhash', 'wfname', 'wfjubedirname', 'wfbenchnr', 'wfrunnr', 'jubedirname', 'benchnr', 'runnr', 'qmom', 'f1bn', 'f1bp', 'f1b', 'f2bnn', 'f2bnp', 'f2bpp', 'f2b', 'norm', 'directr12', 'directradius', 'directnorm', 'rp', 'pnorm', 'rn', 'nnorm', 'LambdaSRGNN', 'LambdaSRGTNF', 'tau4max', 'H[MeV]', 'NORM', 'VNN[MeV]', 'empotsrg', 'j3', 'lsummaxdens', 'mt3', 'tau3max', 'Tmax', 'jobnr', 'jubedir', 'srgwf', 'thetaq']

# first select a few of the columns
colsel=['addtime','kind','N','Z','Jtot','Nmax','OmegaHO','E[MeV]','omega','theta','qmom','thetaq','orderNN','LambdaNN','tnforder','lambdaSRGNN',]

# select for example only n4lo+(450) densities for 4He with omega=80 MeV, no SRG evolution 
omega=135
ts1 = pd.Timestamp("2025-04-26 1:00:00")
select = (
      (df["N"] == 3)
    & (df["LambdaNN"]>10)
    & (df["Z"] == 3)
    & (df["addtime"] >ts1)
    & (df["omega"]<60.1)
    & (df["omega"]>59.9)
    & (df["theta"]==40)
    & (df["LambdaNN"]==450)
    & (df["Nmax"]==10.0)
    & (df["OmegaHO"]==16.0)
    #& (df["lambdaSRGNN"]==1.880)
)

tmpdb=densdb.pddf[select][colsel]
print(len(tmpdb))
tmpdb
```

```python
## download all files consistent with these requirements 
hashlist=list(densdb.pddf[select]['hashname'] )
kindlist=list(densdb.pddf[select]['kind'] )

for n in range(len(hashlist)):
    print("hashname: ",hashlist[n])
    print("kind:     ",kindlist[n])
    # define properties of the file to be downloaded 
    # note that kind has to be specified as well
    fileprops={'hashname':hashlist[n],'kind':kindlist[n]}
    hashfilen,fullfilen=densdb.get_file(**fileprops)
    
    print("hashname: ",hashfilen,"\nfullname: ",fullfilen,"\n\n")
```
